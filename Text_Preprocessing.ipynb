```python
# 2. Remove HTML Tags

import re
def remove_HTMLTags(text):
  pattern = re.compile('<.*?>')
  print(pattern.sub(r'', text))


remove_HTMLTags('<header> <p> </a> Hello There dfjfdj <img> <br> OK')
```

```python
# 3. Remove URLs

def remove_URL(url):
  pattern = re.compile(r'https?://\S+|www\.\S+')
  print(pattern.sub(r'',url))

remove_URL('https://www.google.com sup')
```

```python
# 4. Remove Punctuation

import string

exclude = string.punctuation

def remove_punc(text): #this method is slow
  for char in exclude:
    text = text.replace(char, '')
  print(text)

def remove_punc2(text): #faster method
  print(text.translate(str.maketrans('','', exclude)))

remove_punc('Hi! How is it going?')
remove_punc2('Hi! How is it going?')

```

```python
# 7. Removing Stop Words

# import nltk
# nltk.download('stopwords')
from nltk.corpus import stopwords

def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:] #cloning/copying list
    new_text.clear()
    return " ".join(x)

print(remove_stopwords('Is there anything I can do for you today?'))
```

```python
# 8. Handling Emojis

import re
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

print(remove_emoji("Lmao üòÇüòÇ\n"))

import emoji
print(emoji.demojize("Lmao üòÇüòÇ"))
```

```python
# 10. Stemming - remove inflections in data such as loving loved lovingly all ultimately mean the word "love" so only use the root word

# NLTK has two stemmers - Porter Stemmer (deals mostly with English language) and Snowball Stemmer (deals with all other languages)

from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])

sample = "walk walks walking walked"
stem_words(sample)
```

```python
# 11. Lemmatization - The stem/root word may not be a valid English words (probably has a stem of probabl). To display to a user, we should lemmatise it

# Lemmatization is slower and used only when it is to be shown to user, otherwise Stemmer is better

import nltk
# nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

sentence = "He was running and eating at same time. He has bad habit of swimming after playing long hours in the Sun."
punctuations="?:!.,;"
sentence_words = nltk.word_tokenize(sentence)
for word in sentence_words:
    if word in punctuations:
        sentence_words.remove(word)

sentence_words
print("{0:20}{1:20}".format("Word","Lemma"))
for word in sentence_words:
    print ("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word,pos='v')))
```
